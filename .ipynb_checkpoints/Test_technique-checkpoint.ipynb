| 2 | Extraire, quand elle est présente dans le libellé produit, la **dimension** (ex. `140x190 cm`) et la **couleur** du produit | — |

### 1.3 Qualité des libellés de catégorie : doublons de casse

On vérifie si des catégories `Nature` ne diffèrent que par la casse ou les accents un problème de saisie fréquent qui, sans correction, ferait apparaître deux catégories là où il n'y en a qu'une, et fausserait tout algorithme de classification en aval.

```python
def normaliser_texte(s: str) -> str:
    '''Minuscules + suppression des accents (mais conserve les espaces/ponctuation).'''
    return unicodedata.normalize("NFKD", s).encode("ascii", "ignore").decode().lower().strip()

natures = df["Nature"].dropna().unique()
groupes = {}
for n in natures:
    groupes.setdefault(normaliser_texte(n), []).append(n)

doublons = {k: v for k, v in groupes.items() if len(v) > 1}
print(f"{len(doublons)} groupe(s) de catégories 'Nature' en doublon de casse/accent détecté(s) :")
for k, v in doublons.items():
    print(" -", v)
```

**Constat.** `Défroisseur` / `défroisseur` désignent la même catégorie. C'est un doublon de **forme**, pas de **contenu** : il ne s'agit pas d'une ligne mal catégorisée au sens de l'exercice, mais d'un problème de qualité de référentiel qu'il serait malhonnête d'ignorer. On l'harmonise avant toute chose, car sinon l'algorithme de la partie 3 traiterait ces deux libellés comme deux catégories cibles distinctes.

```python
df["Nature"] = df["Nature"].str.strip()
df.loc[df["Nature"].str.lower() == "défroisseur", "Nature"] = "Défroisseur"
print("Nature distincts après harmonisation :", df["Nature"].nunique())
```

## 2. Partie 1 — Détection et correction des lignes mal catégorisées (`Nature`)

### 2.1 Cadrage du problème

La consigne impose une contrainte forte : **on ne peut pas créer de nouvelle catégorie**. La catégorie correcte d'une ligne mal classée existe forcément déjà quelque part dans la colonne `Nature`. C'est une information précieuse : on n'a pas besoin d'un modèle « ouvert » (topic modeling, clustering libre) mais d'un modèle de **classification fermée** sur les ~595 catégories déjà observées.

### 2.2 Approches envisagées

| Approche | Avantages | Inconvénients | Retenue ? |
|---|---|---|---|
| **Règles manuelles / dictionnaire de mots-clés** (si "matelas" dans le texte → Nature = Matelas) | Simple, explicable, précis | Ne passe pas à l'échelle (595 catégories), demande une expertise métier exhaustive, fragile aux synonymes | Utilisée en **renfort ciblé**, pas en solution principale |
| **Classifieur supervisé "classique"** (Régression logistique, SVM, Naive Bayes) en supposant que les `Nature` actuelles sont les vraies étiquettes | Performant si les labels sont propres | **Hypothèse fausse ici** : on cherche justement les lignes mal étiquetées → le modèle apprend en partie sur du bruit ; coûteux en mémoire avec 595 classes en *one-vs-rest* dense (testé, provoque des `MemoryError` / temps d'entraînement prohibitif sur la machine cible) | Écartée pour la version *from scratch*, mais l'idée est reprise sous une forme robuste ci-dessous |
| **Classification par plus proche centroïde (Rocchio / Nearest Centroid) sur des vecteurs TF‑IDF, avec nettoyage itératif** | Rapide (pas de descente de gradient), très économe en mémoire (un vecteur moyen par catégorie, pas de matrice n_lignes × n_classes), naturellement limitée à l'ensemble des catégories existantes (contrainte de l'énoncé respectée par construction), facile à auditer (on peut expliquer *pourquoi* une ligne est proposée dans telle catégorie) | Suppose que le texte du libellé est discriminant (vrai ici), sensible à la contamination initiale des centroïdes si une catégorie est majoritairement mal peuplée (traité en 2.6) | **Retenue** |

**Principe retenu — "chaque catégorie a un profil vocabulaire, et on rattache chaque produit au profil le plus proche"** :

1. On transforme chaque libellé produit en vecteur numérique (TF‑IDF).
2. Pour chaque catégorie `Nature` existante, on calcule un **vecteur moyen** ("centroïde") à partir des produits actuellement étiquetés dans cette catégorie.
3. Pour chaque ligne du dataset, on compare son vecteur à **tous** les centroïdes : si un centroïde *autre* que celui de sa catégorie actuelle est significativement plus proche, la ligne est suspectée d'être mal catégorisée, et on lui propose ce centroïde comme nouvelle catégorie.
4. Comme les centroïdes de départ peuvent eux-mêmes être pollués par des lignes mal catégorisées, on **itère** : on retire les lignes suspectes avant de recalculer des centroïdes plus propres (un principe proche du *self-training* / *robust estimation*).
5. Une dernière passe de **corroboration lexicale** (le nom de la catégorie proposée doit apparaître littéralement dans le libellé) filtre les faux positifs résiduels avant toute correction automatique.

Ce choix garantit par construction qu'aucune catégorie nouvelle n'est créée : le modèle ne peut prédire qu'une des `Nature` déjà présentes dans les données.

### 2.3 Vectorisation TF‑IDF (par *hashing*)

**Pourquoi le hachage (`HashingVectorizer`) plutôt qu'un `TfidfVectorizer` classique ?**
Un `TfidfVectorizer` classique doit construire et conserver en mémoire le vocabulaire complet (dictionnaire terme → indice), ce qui devient coûteux avec >500 000 libellés et des bi-grammes. Le `HashingVectorizer` projette directement chaque terme dans un espace de taille fixe via une fonction de hachage, sans jamais matérialiser de vocabulaire : l'empreinte mémoire est constante, indépendante du nombre de documents. Le compromis (quelques collisions de hachage) est négligeable ici avec 2¹⁷ ≈ 131 000 dimensions pour ~600 catégories.

On combine bigrammes + unigrammes (`ngram_range=(1,2)`) pour capter des expressions composées ("meuble à chaussures", "canapé d'angle") et pas seulement des mots isolés.

```python
libelles = df["Libellé produit"].astype(str).to_numpy(dtype=object)
nature = df["Nature"]  # déjà nettoyée en 1.3
nature_arr = nature.to_numpy()
est_etiquete = nature.notna().to_numpy()
est_manquant = ~est_etiquete

hv = HashingVectorizer(
    n_features=N_FEATURES_HASH,
    alternate_sign=False,     # composantes positives : compatible TF-IDF / cosinus classique
    ngram_range=(1, 2),
    norm=None,
    dtype=np.float32,
)
tfidf = TfidfTransformer(sublinear_tf=True)  # amortit l'effet des libellés très répétés

t0 = time.time()
X_lab = tfidf.fit_transform(hv.transform(libelles[est_etiquete]))
X_lab = normalize(X_lab, norm="l2").astype(np.float32).tocsr()
print(f"Vectorisation de {X_lab.shape[0]:,} libellés étiquetés en {time.time()-t0:.1f}s".replace(",", " "))
print("Dimension de l'espace vectoriel :", X_lab.shape[1])
```

### 2.4 Construction des centroïdes de référence par catégorie

Le centroïde d'une catégorie est simplement la **moyenne des vecteurs TF‑IDF** des produits actuellement étiquetés dans cette catégorie, puis renormalisée (norme L2 = 1) pour que la comparaison entre catégories se fasse par similarité cosinus.

*Remarque technique* : on évite ici tout encodage "one-hot dense n_lignes × n_catégories" (qui avait provoqué un `MemoryError` en phase d'exploration avec `ComplementNB`/`sklearn`) en utilisant une matrice indicatrice **creuse**, dont le produit matriciel avec `X_lab` donne directement la somme des vecteurs par catégorie.

```python
def construire_centroides(X, y, classes, centroides_repli=None):
    '''Calcule le centroïde TF-IDF (normalisé L2) de chaque catégorie de `classes`
    à partir des lignes (X, y). Si une catégorie n'a aucune ligne dans (X, y),
    on retombe sur `centroides_repli` (utile lors des itérations de nettoyage).'''
    le = LabelEncoder()
    le.fit(classes)
    y_enc = le.transform(y)
    n_classes = len(le.classes_)

    indicatrice = sp.csr_matrix(
        (np.ones(len(y_enc), dtype=np.float32), (np.arange(len(y_enc)), y_enc)),
        shape=(len(y_enc), n_classes),
    )
    sommes = (indicatrice.T @ X).toarray()
    effectifs = np.asarray(indicatrice.sum(axis=0)).ravel()

    centroides = np.divide(sommes, effectifs[:, None], out=np.zeros_like(sommes), where=effectifs[:, None] > 0)
    vides = effectifs == 0
    if vides.any() and centroides_repli is not None:
        centroides[vides] = centroides_repli[vides]
    centroides = normalize(centroides, norm="l2").astype(np.float32)
    return centroides, le.classes_, effectifs


CLASSES = np.sort(nature.dropna().unique())
centroides, classes, effectifs = construire_centroides(X_lab, nature_arr[est_etiquete], CLASSES)
print(f"{len(classes)} centroïdes construits (un par catégorie 'Nature' existante).")
print("Taille moyenne d'une catégorie :", int(effectifs.mean()), "lignes ; médiane :", int(np.median(effectifs)))
```

### 2.5 Score de similarité et détection (1ʳᵉ passe)

Pour chaque ligne on calcule sa similarité cosinus à **tous** les centroïdes (produit matriciel creux × dense, très rapide), puis on retient :
- `best_sim` / `best_idx` : la meilleure catégorie candidate et son score ;
- `cur_sim` : le score de similarité avec la catégorie **actuellement** assignée (pour comparaison).

Le calcul est fait **par blocs** (`chunk`) pour ne jamais matérialiser une matrice dense `n_lignes × n_catégories` complète en mémoire d'un coup (525 034 × 595 float32 ≈ 1,2 Go — gérable en une fois, mais on préfère une marge de sécurité sur une machine à 4 Go de RAM).

```python
def scorer_toutes_les_lignes(centroides, classes):
    '''Calcule pour chaque ligne du dataset la meilleure catégorie candidate
    (similarité cosinus maximale) et la similarité avec sa catégorie actuelle.'''
    classe_to_idx = {c: i for i, c in enumerate(classes)}
    n = len(df)
    best_idx = np.zeros(n, dtype=np.int32)
    best_sim = np.zeros(n, dtype=np.float32)
    cur_sim = np.full(n, np.nan, dtype=np.float32)

    chunk = 40_000
    for start in range(0, n, chunk):
        end = min(start + chunk, n)
        Xc = normalize(tfidf.transform(hv.transform(libelles[start:end])), norm="l2").astype(np.float32)
        sims = np.asarray(Xc @ centroides.T)
        bi = sims.argmax(axis=1)
        best_idx[start:end] = bi
        best_sim[start:end] = sims[np.arange(sims.shape[0]), bi]

        labs = nature_arr[start:end]
        idxs = [classe_to_idx.get(l, -1) if isinstance(l, str) else -1 for l in labs]
        for k, ci in enumerate(idxs):
            if ci >= 0:
                cur_sim[start:end][k] = sims[k, ci]
    return best_idx, best_sim, cur_sim


t0 = time.time()
best_idx, best_sim, cur_sim = scorer_toutes_les_lignes(centroides, classes)
pred_nature = classes[best_idx]
print(f"Scoring de {len(df):,} lignes en {time.time()-t0:.1f}s".replace(",", " "))

# règle de décision (1re passe, avant nettoyage robuste — cf. 2.6)
suspect_mask = est_etiquete & (pred_nature != nature_arr) \
               & (best_sim - np.nan_to_num(cur_sim, nan=-1) > MARGIN) \
               & (best_sim > MIN_CONF)
print(f"Lignes suspectées mal catégorisées (1re passe) : {suspect_mask.sum():,} "
      f"soit {suspect_mask.sum()/est_etiquete.sum()*100:.2f}% des lignes étiquetées".replace(",", " "))
```

### 2.6 Étude de cas : preuve que l'algorithme détecte de vraies erreurs

Avant d'aller plus loin, il est indispensable de **vérifier manuellement** un échantillon de ce que l'algorithme propose : un score de similarité élevé ne prouve rien en soi si on ne confronte pas les résultats à la réalité métier. On inspecte les paires (catégorie actuelle → catégorie proposée) les plus fréquentes parmi les lignes suspectées.

```python
paires = (
    pd.DataFrame({"Nature (actuelle)": nature_arr[suspect_mask], "Nature (proposée)": pred_nature[suspect_mask]})
    .value_counts()
    .head(10)
    .rename("nb_lignes")
)
paires
```

```python
# Vérification manuelle du cas le plus flagrant : "Friteuse" -> "Canapé d'angle"
echantillon = df.loc[(nature_arr == "Friteuse") & (pred_nature == "Canapé d'angle"), "Libellé produit"]
print(f"{(nature_arr == 'Friteuse').sum()} lignes étiquetées 'Friteuse' au total.")
print(f"{echantillon.shape[0]} d'entre elles sont proposées en 'Canapé d'angle'. Échantillon :\n")
print(echantillon.sample(8, random_state=RANDOM_STATE).to_string(index=False))
```

**Verdict.** Sans ambiguïté : des canapés d'angle sont bien étiquetés `Nature = Friteuse` dans le fichier source. On peut même quantifier l'ampleur du problème avec une simple recherche de mot-clé, indépendamment du modèle, pour confirmer objectivement :

```python
friteuse = df[nature_arr == "Friteuse"]
contient_canape = friteuse["Libellé produit"].astype(str).str.lower().str.contains("canap").sum()
print(f"Sur {len(friteuse)} lignes étiquetées 'Friteuse', {contient_canape} "
      f"({contient_canape/len(friteuse)*100:.0f}%) contiennent le mot 'canapé' dans leur libellé.")
```

Cette catégorie `Friteuse` est donc **majoritairement contaminée** par des canapés d'angle — vraisemblablement un bug d'export ou de mapping d'identifiants côté source, en amont du fichier reçu. C'est exactement le type d'anomalie que l'énoncé demande de détecter et corriger.

**Conséquence méthodologique importante.** Si `Friteuse` est peuplée à 88 % de canapés, alors le *centroïde* de `Friteuse` calculé en 2.4 ne représente pas vraiment des friteuses : il représente essentiellement... des canapés. C'est un problème classique d'estimation robuste : **une moyenne calculée sur des données bruitées est elle-même biaisée**. On corrige cela par nettoyage itératif ci-dessous.

### 2.7 Nettoyage itératif des centroïdes (estimation robuste)

**Principe.** On répète le cycle *(scorer → identifier les lignes suspectes → recalculer les centroïdes sans elles)* plusieurs fois. À chaque itération, les catégories fortement contaminées (comme `Friteuse`) perdent leurs intrus, et leur centroïde se rapproche de leur véritable profil vocabulaire. On observe la convergence du nombre de lignes suspectées.

```python
centroides_courants = centroides
historique = []

for iteration in range(N_ITER_REFINE):
    t0 = time.time()
    best_idx, best_sim, cur_sim = scorer_toutes_les_lignes(centroides_courants, classes)
    pred_nature = classes[best_idx]

    suspect_mask = est_etiquete & (pred_nature != nature_arr) \
                   & (best_sim - np.nan_to_num(cur_sim, nan=-1) > MARGIN) \
                   & (best_sim > MIN_CONF)
    propre_mask = est_etiquete & ~suspect_mask

    historique.append({"iteration": iteration, "lignes_suspectes": int(suspect_mask.sum()),
                        "temps_s": round(time.time() - t0, 1)})
    print(f"Itération {iteration} : {suspect_mask.sum():,} lignes suspectes "
          f"(base d'entraînement propre : {propre_mask.sum():,} lignes) — {time.time()-t0:.1f}s".replace(",", " "))

    X_propre = normalize(tfidf.transform(hv.transform(libelles[propre_mask])), norm="l2").astype(np.float32).tocsr()
    centroides_courants, classes, effectifs = construire_centroides(
        X_propre, nature_arr[propre_mask], classes, centroides_repli=centroides_courants
    )

pd.DataFrame(historique)
```

**Lecture du tableau.** Le nombre de lignes suspectes se stabilise après 2 à 3 itérations : le processus converge, signe que les centroïdes ne « dérivent » pas indéfiniment. On conserve les centroïdes de la dernière itération comme référence finale.

Vérifions que la contamination initialement détectée en 2.6 s'est bien résorbée dans les données d'entraînement des centroïdes :

```python
best_idx, best_sim, cur_sim = scorer_toutes_les_lignes(centroides_courants, classes)
pred_nature = classes[best_idx]

mc = nature_arr == "Meuble à chaussures"
print("'Meuble à chaussures' toujours prédites comme telles :",
      f"{(pred_nature[mc]=='Meuble à chaussures').sum()} / {mc.sum()}")
print("'Meuble à chaussures' redirigées vers 'Matelas'          :",
      f"{(pred_nature[mc]=='Matelas').sum()} / {mc.sum()}",
      "  (cohérent avec la contamination croisée observée en exploration)")
```

### 2.8 Renfort par corroboration lexicale — éliminer les faux positifs

Le nettoyage itératif réduit la contamination des centroïdes mais ne l'élimine pas totalement lorsque deux catégories sont *massivement* mélangées dès le départ (`Matelas` ↔ `Meuble à chaussures`, cf. exploration : 68 % des lignes `Meuble à chaussures` mentionnent en réalité « matelas »). Dans ce cas, le signal statistique seul reste ambigu.

**Idée simple et puissante :** le **nom de la catégorie** est lui-même un mot-clé métier très fiable. On ne valide donc une correction *(catégorie actuelle → catégorie proposée)* que si :

1. le(s) mot(s)-clé(s) significatif(s) du nom de la catégorie **proposée** apparaissent littéralement dans le libellé, **ET**
2. le(s) mot(s)-clé(s) de la catégorie **actuelle** n'y apparaissent **pas**.

Si les deux conditions ne sont pas réunies : soit la correction est rejetée (le score de similarité n'était pas corroboré par le texte — probablement un faux positif), soit le cas est **ambigu** (les deux mots-clés sont présents — ex. un lot "canapé + pouf" mentionnant plusieurs types de meubles) et il est alors envoyé en **relecture manuelle** plutôt que corrigé automatiquement à l'aveugle.

```python
def normaliser_pour_motclef(s: str) -> str:
    s = unicodedata.normalize("NFKD", s).encode("ascii", "ignore").decode().lower()
    s = re.sub(r"[^a-z0-9]+", " ", s)
    return " " + s + " "

MOTS_VIDES = {"de", "a", "le", "la", "les", "du", "des", "et", "en", "au", "aux", "d", "par", "pour"}

def mots_cles_categorie(nom_categorie: str):
    tokens = normaliser_pour_motclef(nom_categorie).strip().split()
    return [t for t in tokens if t not in MOTS_VIDES and len(t) > 2]


libelles_normalises = np.array([normaliser_pour_motclef(s) for s in libelles])
cache_motscles = {c: mots_cles_categorie(c) for c in np.unique(np.concatenate([nature_arr[est_etiquete], pred_nature]))
                  if isinstance(c, str)}

def motcle_present(indices, categories):
    resultat = np.zeros(len(indices), dtype=bool)
    for i, idx in enumerate(indices):
        mots = cache_motscles.get(categories[idx], [])
        if not mots:
            continue
        texte = libelles_normalises[idx]
        resultat[i] = all((f" {m} " in texte) or texte.startswith(m + " ") for m in mots)
    return resultat


indices_suspects = np.where(suspect_mask)[0]
motcle_propose_ok = motcle_present(indices_suspects, pred_nature)
motcle_actuel_ok = motcle_present(indices_suspects, nature_arr)

confirmees = motcle_propose_ok & ~motcle_actuel_ok
ambigues = motcle_propose_ok & motcle_actuel_ok
rejetees = ~motcle_propose_ok

print(f"Candidats après nettoyage itératif      : {len(indices_suspects):,}".replace(",", " "))
print(f"  -> Corrections confirmées (mot-clé OK)  : {confirmees.sum():,}".replace(",", " "))
print(f"  -> À vérifier manuellement (ambiguës)    : {ambigues.sum():,}".replace(",", " "))
print(f"  -> Rejetées (non corroborées -> inchangé): {rejetees.sum():,}".replace(",", " "))
```

Vérifions que ce garde-fou lexical tranche correctement les deux cas emblématiques identifiés pendant l'exploration :

```python
masque_correction = np.zeros(len(df), dtype=bool)
masque_correction[indices_suspects[confirmees]] = True
masque_a_verifier = np.zeros(len(df), dtype=bool)
masque_a_verifier[indices_suspects[ambigues]] = True

for cat_actuelle, cat_proposee in [("Matelas", "Meuble à chaussures"), ("Meuble à chaussures", "Matelas")]:
    m = (nature_arr == cat_actuelle) & (pred_nature == cat_proposee) & masque_correction
    print(f"{cat_actuelle:22s} -> {cat_proposee:22s} : {m.sum()} corrections finalement confirmées")
```

**Résultat.** Le sens `Matelas → Meuble à chaussures` (qui aurait été un **faux positif** : ces lignes sont de vrais matelas déjà bien étiquetés, seulement "proches" du centroïde encore un peu contaminé de `Meuble à chaussures`) est intégralement rejeté. Le sens inverse, `Meuble à chaussures → Matelas` (de vrais matelas mal étiquetés), reste confirmé. La corroboration lexicale agit exactement comme prévu : c'est un filtre à haute précision, complémentaire du signal statistique.

### 2.9 Traitement des `Nature` manquantes

On applique exactement la même logique aux ~11 700 lignes sans `Nature` : proposer la catégorie la plus proche, mais ne l'assigner automatiquement que si le mot-clé de la catégorie proposée est bien présent dans le libellé.

```python
indices_manquants = np.where(est_manquant & (best_sim > MIN_CONF))[0]
manquant_ok = motcle_present(indices_manquants, pred_nature)

masque_remplissage = np.zeros(len(df), dtype=bool)
masque_remplissage[indices_manquants[manquant_ok]] = True

print(f"Lignes à 'Nature' manquante           : {est_manquant.sum():,}".replace(",", " "))
print(f"  -> candidates (similarité > seuil)   : {len(indices_manquants):,}".replace(",", " "))
print(f"  -> confirmées par mot-clé (remplies) : {manquant_ok.sum()}")
print()
print("Échantillon de lignes à Nature manquante, non comblées automatiquement :")
df.loc[est_manquant, "Libellé produit"].sample(8, random_state=RANDOM_STATE).to_frame()
```

**Constat honnête.** Le remplissage automatique des `Nature` manquantes ne concerne finalement aucune ligne avec ce niveau d'exigence (aucun mot-clé corroborant trouvé sur l'échantillon candidat). En creusant, on comprend pourquoi : une bonne partie de ces lignes ne sont pas de vrais produits d'ameublement mais des **lignes de service** (abonnement, assistance téléphonique, programme de fidélité…) qui n'ont *aucune* catégorie `Nature` pertinente parmi celles existantes — et il serait incorrect de leur en forcer une.

C'est un résultat volontairement conservateur : **mieux vaut ne rien proposer que proposer une catégorie non fiable**. Ces lignes restent marquées « à vérifier manuellement », ce qui est le comportement responsable attendu d'un outil de correction automatique en production.

### 2.10 Synthèse des corrections & validation qualitative finale

```python
statut = np.full(len(df), "OK (inchangée)", dtype=object)
nature_corrigee = nature_arr.copy()

nature_corrigee[masque_correction] = pred_nature[masque_correction]
statut[masque_correction] = "Corrigée automatiquement"
statut[masque_a_verifier] = "Ambiguë - à vérifier manuellement"
statut[est_manquant] = "Nature manquante - à vérifier manuellement"

synthese = pd.Series(statut).value_counts().to_frame("nb_lignes")
synthese["% du total"] = (synthese["nb_lignes"] / len(df) * 100).round(2)
synthese
```

### 2.11 Mesure de performance indicative (train / test)

À titre de contrôle complémentaire (et non comme preuve suffisante à elle seule, car les étiquettes de test sont elles-mêmes potentiellement bruitées), on mesure la capacité du modèle de plus-proche-centroïde à retrouver l'étiquette actuelle sur un jeu de test mis de côté. Un score élevé confirme que le texte du libellé est bien discriminant pour prédire la catégorie.

```python
X_idx_train, X_idx_test, y_train, y_test = train_test_split(
    np.arange(est_etiquete.sum()), nature_arr[est_etiquete], test_size=0.2, random_state=RANDOM_STATE
)
lib_lab = libelles[est_etiquete]

Xtr = normalize(tfidf.transform(hv.transform(lib_lab[X_idx_train])), norm="l2").astype(np.float32).tocsr()
centroides_eval, classes_eval, _ = construire_centroides(Xtr, y_train, CLASSES)

Xte = normalize(tfidf.transform(hv.transform(lib_lab[X_idx_test])), norm="l2").astype(np.float32)
pred_test = classes_eval[np.asarray(Xte @ centroides_eval.T).argmax(axis=1)]

acc = accuracy_score(y_test, pred_test)
print(f"Exactitude (accuracy) sur le jeu de test, vs. étiquette actuelle (potentiellement bruitée) : {acc*100:.1f}%")
print("-> Un score élevé malgré le bruit d'étiquetage confirme que le texte du libellé est bien",
      "discriminant, ce qui valide l'approche retenue.")
```

### 2.12 Limites de l'approche & pistes d'amélioration

Il est important, dans une correction de référence, d'assumer explicitement les limites de la méthode plutôt que de laisser croire à une solution parfaite :

- **Seuils `MARGIN` / `MIN_CONF` fixés empiriquement.** Ils ont été choisis par observation itérative des résultats plutôt qu'optimisés sur une vérité terrain labellisée (qui n'existe pas ici). Une évaluation plus rigoureuse nécessiterait un échantillon annoté manuellement pour calculer précision/rappel du dispositif de correction.
- **Catégories très majoritairement contaminées dès le départ** (ex. `Meuble à chaussures` à 68 % de matelas) : le nettoyage itératif réduit fortement le biais mais la garde lexicale, volontairement stricte, laisse une partie des lignes réellement mal catégorisées dans le statut "inchangée" plutôt que de risquer une correction hasardeuse (choix délibéré : **précision > rappel**).
- **Catégories rares** (moins de 5 à 10 lignes) : le centroïde associé est estimé sur très peu d'exemples et donc statistiquement fragile. Une v2 pourrait fusionner ces micro-catégories avec un a priori métier, ou les exclure du périmètre de correction automatique.
- **La corroboration lexicale suppose que le nom de la catégorie apparaît tel quel dans le texte.** Ce n'est pas toujours vrai (ex. "clic-clac" pour "Canapé convertible"). Un dictionnaire de synonymes métier, construit avec les équipes catalogue, améliorerait le rappel sans sacrifier la précision.
- **Bundles / produits composites** ("lit + matelas + sommier") créent des ambiguïtés réelles où plusieurs catégories sont légitimement mentionnées : le statut "ambiguë" est adapté à ce cas mais suppose une relecture humaine.

**Pour aller plus loin (hors périmètre du test, à discuter à l'oral)** : un modèle de langue pré-entraîné (embeddings de type FastText / CamemBERT) capturerait mieux les synonymes et fautes de frappe qu'une simple pondération TF‑IDF ; un tableau de bord de contrôle qualité continu permettrait de détecter automatiquement les nouvelles dérives de catégorisation à chaque import de données.

## 3. Partie 2 — Extraction de la dimension et de la couleur

### 3.1 Méthodologie

Contrairement à la catégorisation (partie 2), extraire une dimension ou une couleur est une tâche de **reconnaissance de motifs dans le texte**, pas une tâche de classification statistique : une expression régulière bien conçue, doublée d'un dictionnaire métier pour la couleur, est l'outil le plus adapté plus précis, plus rapide et bien plus explicable qu'un modèle statistique pour ce type d'extraction structurée.

**Dimension.** On cherche des motifs du type `L x l (x H) [unité]`, par exemple :
`140x190 cm`, `90 x 190`, `120x40x98 cm`, `80l x 23,5l x 124h cm`…

**Couleur.** On s'appuie sur un dictionnaire de couleurs et teintes usuelles dans l'ameublement (couleurs pures + teintes bois type « chêne », « noyer »), avec priorité aux expressions composées (`gris anthracite` avant `gris` seul).

### 3.2 Extraction de la dimension

```python
NUM = r"\d{1,4}(?:[.,]\d+)?"
SUFFIXE_LETTRE = r"(?:[lLhHpPdD](?![a-zA-Z]))?"   # ex. le 'l'/'h'/'p' de 80l x 23l x 124h (Longueur/Hauteur/Profondeur)
UNITE = r"\s*(cm|mm|m)?\b"

RE_DIMENSION = re.compile(
    rf"({NUM}){SUFFIXE_LETTRE}\s*[xX×\*]\s*({NUM}){SUFFIXE_LETTRE}\s*(?:[xX×\*]\s*({NUM}){SUFFIXE_LETTRE})?{UNITE}"
)

def extraire_dimension(libelle: str):
    '''Renvoie la dimension trouvée sous forme 'L x l(x H) unité', ou None.
    S'il y a plusieurs motifs candidats dans le libellé (ex. un nombre de lattes
    puis une vraie dimension), on privilégie celui qui est suivi d'une unité
    explicite (cm/mm/m), plus fiable.'''
    candidats = list(RE_DIMENSION.finditer(libelle))
    if not candidats:
        return None
    avec_unite = [m for m in candidats if m.group(4)]
    choisi = avec_unite[0] if avec_unite else candidats[0]
    g = choisi.groups()
    dims = [g[0].replace(",", "."), g[1].replace(",", ".")]
    if g[2]:
        dims.append(g[2].replace(",", "."))
    unite = g[3] or "cm"  # hypothèse documentée : l'unité par défaut du secteur ameublement est le cm
    return "x".join(dims) + " " + unite


# Vérification sur l'exemple donné dans l'énoncé
exemple = "Matelas mousse 140x190 cm"
print(f"Exemple énoncé : {exemple!r} -> {extraire_dimension(exemple)!r}")
```

```python
t0 = time.time()
dimension = df["Libellé produit"].astype(str).apply(extraire_dimension)
print(f"Extraction sur {len(df):,} lignes en {time.time()-t0:.1f}s".replace(",", " "))
taux = dimension.notna().mean() * 100
print(f"Dimension détectée sur {dimension.notna().sum():,} lignes ({taux:.1f}% du dataset)".replace(",", " "))
```

```python
pd.DataFrame({"Libellé produit": df["Libellé produit"], "Dimension extraite": dimension})[dimension.notna()].sample(10, random_state=RANDOM_STATE)
```

**Interprétation du taux de couverture (30 %).** Ce chiffre n'est pas une anomalie : une large part du catalogue (chaises, fauteuils, petit électroménager par exemple) ne mentionne tout simplement pas de dimension dans son libellé commercial. Le taux de couverture doit donc être lu catégorie par catégorie plutôt que globalement :

```python
(pd.DataFrame({"Nature": nature_corrigee, "a_dimension": dimension.notna()})
 .groupby("Nature")["a_dimension"].mean().sort_values(ascending=False).head(10) * 100
).round(1).to_frame("% de lignes avec dimension détectée")
```

Sans surprise, ce sont les catégories où la dimension est un critère d'achat déterminant (matelas, sommiers, tapis, rideaux, cadres à lattes…) qui affichent les meilleurs taux — ce qui confirme la cohérence du dispositif.

### 3.3 Extraction de la couleur

On combine :
1. une normalisation du texte (minuscule, suppression des accents) qui **corrige aussi les artefacts d'encodage identifiés en 1.4** (`ch ne` → `chene`, `h tre` → `hetre`) pour ne pas perdre ces occurrences ;
2. un dictionnaire de couleurs/teintes usuelles du secteur, trié par longueur décroissante pour que les expressions composées (`gris anthracite`) soient reconnues avant leur composante simple (`gris`).

```python
def normaliser_couleur(s: str) -> str:
    s = unicodedata.normalize("NFKD", s).encode("ascii", "ignore").decode().lower()
    s = re.sub(r"\bch ne\b", "chene", s)
    s = re.sub(r"\bh tre\b", "hetre", s)
    s = re.sub(r"\bt te\b", "tete", s)
    s = re.sub(r"\bcr me\b", "creme", s)
    return s

COULEURS_REFERENCE = [
    "blanc", "noir", "gris anthracite", "gris souris", "gris perle", "gris clair", "gris fonce", "gris",
    "beige", "marron", "taupe", "naturel", "chene clair", "chene fonce", "chene", "hetre", "noyer", "wenge",
    "bleu marine", "bleu ciel", "bleu petrole", "bleu", "vert olive", "vert amande", "vert sapin", "vert",
    "rouge", "rose poudre", "rose gold", "rose", "jaune moutarde", "jaune", "orange", "violet", "prune",
    "argente", "argent", "dore", "or", "cuivre", "bronze", "ivoire", "ecru", "kaki", "turquoise", "corail",
    "moutarde", "bordeaux", "cognac", "sable", "camel", "chocolat", "creme", "ocre", "terracotta",
    "multicolore", "transparent", "fuchsia", "lin", "pastel", "nude", "anthracite", "noir mat", "blanc mat",
    "blanc laque", "noir laque", "blanc brillant", "noir brillant", "gris beton", "marbre blanc", "marbre noir",
]
COULEURS_REFERENCE = sorted(set(COULEURS_REFERENCE), key=lambda x: -len(x))
RE_COULEUR = re.compile(r"\b(" + "|".join(re.escape(c) for c in COULEURS_REFERENCE) + r")\b")

def extraire_couleur(libelle: str):
    m = RE_COULEUR.search(normaliser_couleur(libelle))
    return m.group(1) if m else None

# Vérification sur un exemple avec l'artefact d'encodage
exemple = "Meuble tv scandinave 100 cm blanc mat ch ne clair"
print(f"Exemple : {exemple!r} -> {extraire_couleur(exemple)!r}")
```

```python
t0 = time.time()
couleur = df["Libellé produit"].astype(str).apply(extraire_couleur)
print(f"Extraction sur {len(df):,} lignes en {time.time()-t0:.1f}s".replace(",", " "))
taux = couleur.notna().mean() * 100
print(f"Couleur détectée sur {couleur.notna().sum():,} lignes ({taux:.1f}% du dataset)".replace(",", " "))
couleur.value_counts().head(15).to_frame("nb_lignes")
```

### 3.4 Résultat combiné sur l'exemple de l'énoncé

```python
exemple = "Matelas mousse 140x190 cm"
print(f"Libellé   : {exemple}")
print(f"Dimension : {extraire_dimension(exemple)}")
print(f"Couleur   : {extraire_couleur(exemple)}   (absente ici, comme dans l'exemple original)")
```

### 3.5 Limites de l'extraction dimension / couleur

- **Dimensions au format dégradé** : certains libellés comportent une anomalie de saisie où le séparateur décimal (virgule) semble avoir été remplacé par une espace (ex. `23 5l` pour `23,5l`). Ces cas ne sont pas couverts par la regex actuelle et représentent une piste d'amélioration si le volume le justifie.
- **Unité par défaut supposée = cm** lorsqu'aucune unité n'est explicitement précisée après un motif `LxP`. C'est l'hypothèse dominante dans l'ameublement, mais elle n'est jamais vérifiée pour les libellés sans unité — hypothèse documentée dans le code, comme l'exige la bonne pratique.
- **Dictionnaire de couleurs non exhaustif** : construit par inspection empirique des données (fréquences de mots), il couvre les teintes les plus fréquentes mais pas l'intégralité du champ lexical possible (ex. couleurs de marque très spécifiques). Un enrichissement collaboratif avec les équipes catalogue serait la suite logique.
- **Un seul couleur/dimension par ligne** : si un libellé mentionne plusieurs couleurs (ex. un lot multicolore décrit couleur par couleur), seule la première rencontrée est retenue. Le besoin exprimé dans l'énoncé (une valeur par produit) est cependant respecté.
